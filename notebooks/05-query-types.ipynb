```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as clrs
from collections import OrderedDict
```

```python
results = OrderedDict()


def open_and_load_data(directory):
    files = sorted(os.listdir(directory))
    for filename in files:
        f = open(directory + '/' + filename, 'r')
        this_month = OrderedDict()
        for line in f:
            query_type, count = line.split()
            this_month[query_type] = int(count)
        results[filename] = this_month
        

def plot_lines_for_months():
    x = np.arange(12)
    months = ['January', 'Feburary', 'March', 'April', 'May', 'June', 'July', 
              'August', 'September', 'October', 'November', 'December']
    ns = [result['NS'] for month, result in results.items()]
    soa = [result['SOA'] for month, result in results.items()]
    mx = [result['MX'] for month, result in results.items()]
    a = [result['A'] for month, result in results.items()]
    txt = [result['TXT'] for month, result in results.items()]
    cname = [result['CNAME'] for month, result in results.items()]
    combined = [cname, ns, soa, mx, a, txt]
    labels = ['CNAME', 'NS', 'SOA', 'MX', 'A', 'TXT']
    plt.figure(figsize=(10, 10))
    plt.xticks(x, months, rotation=45)
    for i, r in enumerate(combined):
        ax = plt.subplot()
        ax.plot(r, label=labels[i])
        ax.legend()
```

```python
open_and_load_data('../data/dns_records_analysis_results')
plot_lines_for_months()
plt.show()
```
